```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline  

import seaborn as sns
```

```python
df_15= pd.read_csv('2015.csv', index_col= 'Country') # 158 rows, 12 cols  
# the happiness score is the sum of all the 7 columns after std error

df_16= pd.read_csv('2016.csv', index_col = 'Country')  # 157 rows, 13 col
df_17= pd.read_csv('2017.csv', index_col = 'Country')  # 155 x 11
df_18= pd.read_csv('2018.csv', index_col = 'Country or region')    # 156 x 12
df_19= pd.read_csv('2019.csv', index_col = 'Country or region')    # 156 x 8
df_20= pd.read_excel('2020.xls', index_col = 'Country name')   # 153 x 20
```

```python
# Rename the columns as Factor+Year e.g. Economy2017, Generosity2020, ..

happyness_rank = ['Happiness Rank', 'Happiness Rank', 'Happiness.Rank',
                 'Overall rank', 'Overall rank', 'Happiness Rank']
happyness_score = ['Happiness Score', 'Happiness Score', 'Happiness.Score', 
                  'Score', 'Score', 'Ladder score']
economy = ['Economy (GDP per Capita)', 'Economy (GDP per Capita)',
          'Economy..GDP.per.Capita.', 'GDP per capita', 'GDP per capita',
          'Logged GDP per capita']
family = ['Family', 'Family', 'Family', 
          'Social Support', 'Social Support', 'Social Support']
health = ['Health (Life Expectancy)', 'Health (Life Expectancy)', 
         'Health..Life.Expectancy.', 'Healthy life expectancy', 
         'Healthy life expectancy', 'Healthy life expectancy']
freedom = ['Freedom', 'Freedom', 'Freedom', 
          'Freedom to make life choices', 'Freedom to make life choices', 
          'Freedom to make life choices']
trust = ['Trust (Government Corruption)', 'Trust (Government Corruption)',
         'Trust..Government.Corruption.', 'Perceptions of corruption',
         'Perceptions of corruption', 'Perceptions of corruption']
# Generosity is ok


df = [df_15, df_16, df_17, df_18, df_19, df_20]
factors = [happyness_rank, happyness_score,
    economy, family, health, freedom, trust, ['Generosity' for i in range(6)]]
names = ['Rank', 'Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity']
for i in range(len(df)):
    cols= dict()
    for j in range(len(factors)):
        cols[factors[j][i]] = names[j]+str(2015+i)
    df[i].rename(columns = cols, inplace= True)
```

```python
df_15
```

```python
# function that creates a dataframe with scores/family/economy/.. per year, indexed on country names
# name is the factor: Rank, Family, ...
def dataframe_years(name):
    df = df_15[[name+'2015']]
    df = df.join(df_16[[name+'2016']], how= 'outer')
    df = df.join(df_17[[name+'2017']], how= 'outer')
    df = df.join(df_18[[name+'2018']], how= 'outer')
    df = df.join(df_19[[name+'2019']], how= 'outer')
    df = df.join(df_20[[name+'2020']], how= 'outer')
    return df
```

```python
# scores:

scores = dataframe_years('Score')
scores.head()
```

```python
# Deal with NaNs

scores[scores.isnull().any(axis=1)]
```

```python
# Let's study how happyness changed during the years

scores_delta = scores.copy()

for n in range(5):
    interval = str(15+n)+'-'+str(15+n+1)  # for ex. '15-16' in the first iteration
    scores_delta[interval] = scores_delta['Score'+str(2015+n+1)]-scores_delta['Score'+str(2015+n)] # scores of 2016 minus scores of 2015

scores_delta.drop(labels = ['Score2015',
                      'Score2016',
                      'Score2017',
                      'Score2018',
                      'Score2019',
                      'Score2020'], axis=1,
            inplace= True)
scores_delta.head()
```
